```python
for i in range(len(data)):
    if (cdt-data['TimeStamp'][i]<timedelta(hours=1)):
        data['Status'][i]='ONLINE'
        list_a.append(data['SiteName'][i])
        a=a+1
    else:
        data['Status'][i]='OFFLINE'
        list_b.append(data['SiteName'][i])
        b=b+1
```

```python
for i in range(len(data)):
    if (cdt-data['TimeStamp'][i]>timedelta(hours=1)):
        data['Offline Since'][i]=cdt-data['TimeStamp'][i]
```

```python
for i in range(len(data)):
    if (data['Status'][i]=="ONLINE" and data["ExternalBattery"][i]==False and data["InternalBattery"][i]>36 and data["InternalBattery"][i]<40):
        data["IBV"][i]=int(data["InternalBattery"][i])*0.1
```
